### Part 2.2 : Analysis of Customer Worksheet

#### 1. 	Which age group spent most money on products based on 2017 Transactions?

#### Change the DOB to age

```python
def get_age(df):
    now = datetime.datetime.now()   #returns today's date
    df['age'] = now.year - pd.DatetimeIndex(df['DOB']).year
    df.drop(columns = ["DOB"], inplace = True)
    return df['age']
```

```python
get_age(df_Cust)
df_Cust.head()
```

```python
df_Cust.age.max()
```

```python
def age_to_category(df):
    bins = [18, 25, 35, 60, 100]
    group_names = ['Youth', 'Young-Adult', 'Middle-Aged', 'Senior']
    df['category'] = pd.cut(df['age'], bins, labels = group_names)
    return (df.head())
```

```python
age_to_category(df_Cust)
```

####  3. Number of customers breakdown by gender and industry

```python
df_Cust.head()
```

```python
df5 = df_Cust[['gender','job_industry_category']]
df5.head()
```

```python
print(df5['gender'].unique())
df5['job_industry_category'].unique()
```

```python
df5 = df5.dropna()
```

```python
c1 = df5.gender != 'Undefined'
df5[c1]
```

```python
c3 = df5.gender == 'Male'
male_df = df5[c3]
male_df.head()
```

```python
c4 = df5.gender == 'Female'
female_df = df5[c4]
female_df.head()
```

```python
f_df = female_df.groupby('job_industry_category').count()
f_df.gender.tolist()
```

```python
m_df = male_df.groupby('job_industry_category').count()
m_df.gender.tolist()
```

```python
labels = ['Argi', 'Enter', 'Fin', 'Health', 'IT', 'Manu' , 'Property', 'Retail', 'Telecom']
men_means = m_df['gender']
women_means = f_df['gender']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label = 'Men')
rects2 = ax.bar(x + width/2, women_means, width, label = 'Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('number of customers')
ax.set_title('Customer by Industry & Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
